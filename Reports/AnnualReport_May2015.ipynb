```python
import numpy as np
import pandas as pd
import os
from os import listdir
import sys
import pprint
import csv
```

```python
from pylab import *
import matplotlib.pyplot as plt
import matplotlib.pylab as pylab

%matplotlib inline
```

#### Import Data

```python
#function to find files in defined path
def find_csv_filenames( path_to_dir, suffix=".csv" ):
    filenames = listdir(path_to_dir)
    return [ filename for filename in filenames if filename.endswith( suffix ) ]

def find_immediate_subdirectories(path_to_dir):
    return [name for name in listdir(path_to_dir)
            if os.path.isdir(os.path.join(path_to_dir, name))]
```

#### Define files to be read

```python
path_cal = '/Users/Lampe/GrantNo456417/MODELING/SierraModeling/Gomez_April6-2015/SALTSHAFTCLOSURE_CALLAHAN/ADAGIO_RTEST_MYTESTSANDBOX_SALTCONSOLIDATION_SALT.TEST'
path_lampe = '/Users/Lampe/GrantNo456417/MODELING/SierraModeling/Gomez_April6-2015/saltShaftClosure_Lampe/adagio_rtest_myTestSandbox_saltConsolidation_salt.test'
path = [path_cal, path_lampe]
subdir = []
sub_path = []
cnt = 0
m = 0
n = 0

for i in path: # loop over both directories defined by 'path'
    subdir = find_immediate_subdirectories(i)
    for j in subdir: # loop over each of the subdirectories defined by 'subdir'
        sub_path = i + '/' + j # concatenate 
        filenames = find_csv_filenames(sub_path) #identify .csv files
        print filenames[0]
        full_path = sub_path + '/' + filenames[0] # concatenate to identify the full path
        reader = csv.reader(open(full_path, 'rb'),delimiter = ',') # read data
        x = list(reader)
        colname = x[1] #identify column names
        result = np.array(x[2:len(x)]).astype('float')
        print result.shape

        if m == 0 and n == 0:
            cal_430 = result
        elif m == 0 and n == 1:
            cal_515 = result
        elif m == 0 and n == 2:
            cal_550 = result
        elif m == 0 and n == 3:
            cal_600 = result
        elif m == 1 and n == 0:
            lampe_430 = result
        elif m == 1 and n == 1:
            lampe_515 = result
        elif m == 1 and n == 2:
            lampe_550 = result
        elif m == 1 and n == 3:
            lampe_600 = result
            
        cnt = cnt + 1
        n = n + 1
    m = m + 1
    n = 0
print colname
```

#### Influence of Water Content at SHEAR Stress at 30 C

```python
fig_reldenAv, ax = plt.subplots(figsize = (12,8))

y = 3
x = 2

print colname[x]
print colname[y]

ax.plot(cal_430[:,x], cal_430[:, y], 'r-',markersize = 4, lw = 2, label=''r' Callahan: 430 m')
ax.plot(cal_515[:,x], cal_515[:, y], 'b-',markersize = 4, lw = 2, label=''r' Callahan: 515 m')
ax.plot(cal_550[:,x], cal_550[:, y], 'g-',markersize = 4, lw = 2, label=''r' Callahan: 550 m')
ax.plot(cal_600[:,x], cal_600[:, y], 'c-',markersize = 4, lw = 2, label=''r' Callahan: 600 m')

ax.plot(lampe_430[:,x], lampe_430[:, y], 'r--',markersize = 4, lw = 2, label=''r' Lampe: 430 m')
ax.plot(lampe_515[:,x], lampe_515[:, y], 'b--',markersize = 4, lw = 2, label=''r' Lampe: 515 m')
ax.plot(lampe_550[:,x], lampe_550[:, y], 'g--',markersize = 4, lw = 2, label=''r' Lampe: 550 m')
ax.plot(lampe_600[:,x], lampe_600[:, y], 'c--',markersize = 4, lw = 2, label=''r' Lampe: 600 m')

ax.legend(loc=0, fontsize = 12)
ax.set_ylim(0.9, 1.0)
# ax.set_xscale('log')
# ax.set_yscale('log')
plt.rcParams['font.size']=20
ax.tick_params(axis = 'x', labelsize = 20)
ax.tick_params(axis = 'y', labelsize = 20)

ax.set_xlabel(''r'Time [years]', fontsize = 25)
ax.set_ylabel(''r'Fractional Density', fontsize = 25)
ax.set_title(''r'Average Fractional Density Predicted During Shaft Closure', fontsize = 20)
ax.grid(b = True, which = 'minor')
ax.grid(b = True, which = 'major')
fig_reldenAv.savefig("/Users/Lampe/GrantNo456417/DOE_Reports/AnnualReprt_May2015/Figures/LampeVsCal_fDen-time.jpg")
```
