```python
import numpy as np
import pandas as pd
from os import listdir
import sys
import pprint
import csv
```

```python
from pylab import *
import matplotlib.pyplot as plt
import matplotlib.pylab as pylab

%matplotlib inline
```

#### Import Data

```python
#function to find files in defined path
def find_csv_filenames( path_to_dir, suffix=".csv" ):
    filenames = listdir(path_to_dir)
    return [ filename for filename in filenames if filename.endswith( suffix ) ]
```

#### Influence of Water Content at SHEAR Stress at 30 C

```python
path = "/Users/Lampe/GrantNo456417/DOE_Reports/MilestoneReport/May2015/R_modeling/SH_30"
run_name = []
filenames = find_csv_filenames(path)
data_count = len(filenames)

out = np.zeros((230, 24, data_count))
print out.shape
colname = []


i = 0
for name in filenames:
    print name
    path_name = path + '/' + name
    reader = csv.reader(open(path_name, 'rb'),delimiter = ',')
    x = list(reader)
    colname = x[0]
    del x[0]
    result = np.array(x).astype('float')
    out[:,:,i] = result
    i = i + 1
    
print colname
```

```python
fig_fden, ax = plt.subplots(figsize = (12,8))

ax.plot(out[0:175,0,0], out[0:175,22,0], 'r-',markersize = 4, lw = 2, label=''r' w = 0% (Recent)')
ax.plot(out[0:175,0,1], out[0:175,22,1], 'b-',markersize = 4, lw = 2, label=''r' w = 0.01% (Recent)')
ax.plot(out[0:175,0,2], out[0:175,22,2], 'g-',markersize = 4, lw = 2, label=''r' w = 0.10% (Recent)')
ax.plot(out[0:175,0,3], out[0:175,22,3], 'c-',markersize = 4, lw = 2, label=''r' w = 1.00% (Recent)')
ax.plot(out[0:175,0,4], out[0:175,22,4], 'm-',markersize = 4, lw = 2, label=''r' w = 10.0% (Recent)')

ax.plot(out[0:175,0,5], out[0:175,22,5], 'r--',markersize = 4, lw = 2, label=''r' w = 0% (Original)') # Callahn Parameters
ax.plot(out[0:175,0,6], out[0:175,22,6], 'b--',markersize = 4, lw = 2, label=''r' w = 0.01% (Original)')
ax.plot(out[0:175,0,7], out[0:175,22,7], 'g--',markersize = 4, lw = 2, label=''r' w = 0.10% (Original)')
ax.plot(out[0:175,0,8], out[0:175,22,8], 'c--',markersize = 4, lw = 2, label=''r' w = 1.00% (Original)')
ax.plot(out[0:175,0,9], out[0:175,22,9], 'm--',markersize = 4, lw = 2, label=''r' w = 10.0% (Original)')

ax.legend(loc=0, fontsize = 12)
ax.set_ylim(0.9, 1.0)
# ax.set_xscale('log')
# ax.set_yscale('log')
plt.rcParams['font.size']=20
ax.tick_params(axis = 'x', labelsize = 20)
ax.tick_params(axis = 'y', labelsize = 20)

ax.set_xlabel(''r'Time [seconds]', fontsize = 25)
ax.set_ylabel(''r'Fractional Density', fontsize = 25)
ax.set_title(''r'Shear Stress of 20 MPa Axial and 10 MPa Lateral at 30$\degree$ C', fontsize = 20)
ax.grid(b = True, which = 'minor')
ax.grid(b = True, which = 'major')
fig_fden.savefig("/Users/Lampe/GrantNo456417/DOE_Reports/MilestoneReport/May2015/Figures/sh_water_content_30C.jpg")
```
