# Pressure Falloff Permeability Test
Reference: *API RP40 Section 6.4.1.1*

## Transient Analysis Method
A numerical (approximate) method is used to obtain a solution to this transient problem. 
The steady-state solution of the slip-corrected Forchheimer equation is used as an initial approximation.  The initial approximation does not account for a change in mass flux the sample length.  Therefore, the initial approximation is differentiated (to obtain the rate of change in mass flux over the sample length) and then inserted into the continuity equation. The continuity equation is then integrated over the sample length to obtain an improved approximation to the flow equation.  This process is iteratively repeated until bth the Forchheimer and continuity equations are satisfied within a given tolerance.
The deviate between steady-state and pressure-falloff solutions depends upon the ratio of the sample's pore volume to the volume of th eupstream gas reservoir.  When this ratio is small, the steady-state solution is nearly exact.

## Precautions

**Thermal effects** for high permeability samples use:
    - low pressure (<25 psi)
    - small diameter (~1 inch)
    - long sample lengths (>2 inches)
    
**Long measurement time** for low permeability samples use:
    - large diameter
    - short sample lengths
    
**Stress sensitivite results** are often observed in low permeability samples, to minimize use:
    - high confining stress (>2000 psi)
    - limit variation of pore pressure during test

```python
c1 = 1.0
c2 = 1.0e12
c3 = 1.0
c4 = 1.0
R = 8314.0 #gas const (m3-Pa)/(K-kmol)
g = 9.81 # gravitational acceleration (m/s2)

def visc_n2(T,P):
#     """
#     viscosity of N2 at ~25 deg C for pressure up to 3700 psi
#     - from API RP40 pg 6-8, Table 6-5 (Viscosity of Nitrogen)
#     - returns viscosity in units of Pa-s
#     """
#     p_atm = p_psi / 14.6959 # gauge pressure in atmospheres
#     Patm_atm = Patm_psi / 14.6959 # atmospheric pressure in atmospheres
#     P = p_atm + Patm_atm # absolute pressure in atmospheres
#     T = 25 + 273.15
    
#     mu1 = (13.85*T**1.5)/(T+102)# Viscosity of N2 (micropoises) at 1 atm
#     v1 = -0.12474
#     v2 = 0.123688
#     v3 = 1.05452e-3
#     v4 = -1.5052e-6
#     mu = (mu1 + v1 + v2*P + v3*P**2 + v4*P**3)*1.0e-7 #viscosity
    """
    T = nitrogen temperature in K
    P = absolute pressure in Pa
    """
    mu = PropsSI('viscosity', 'T', T, 'P', P, 'Nitrogen') # Pa-s
    return mu

def z_n2(T,P):
    """
    T = nitrogen temperature in K
    P = absolute pressure in Pa
    """
    # Density of Air at standard atmosphere in kg/m^3
    z = PropsSI('Z','T',T,'P',P,'Nitrogen')
    return z

def rho_n2(T,P):
    """
    returns density of nitrogen (kg/m3)
    T = nitrogen temperature in K
    P = absolute pressure in Pa
    """
    M = 28.01348 #molecular weight of nitrogen (kg/kg-mole)
    R = 8314.0 #gas const (m3-Pa)/(K-kmol)
    z = z_n2(T,P)
    rho = M*P/(z*R*T)
#     PropsSI('D', 'T', 298.15, 'P', 101325, 'Nitrogen')
    return rho
```

```python
T = T_c + 273.15 # abs temp (K)
L = (L_in/12.0)*0.3047 # length (m)
D = (d_in/12.0)*0.3047 # diam (m)
A = (np.pi * D**2)/4.0 # area of sample orth to flow direction (m2)
M = 28.0134 # molecular weight of nitrogen (kg/kg-mole)
Patm_pa = Patm_psi*(101.325e3 / 14.6959) # downstream absolute gas pressure (Pa)
p_pa = p_psi*(101.325e3 / 14.6959) # upstream gauge gas pressure (Pa)
P_pa = (p_psi + Patm_psi)*(101.325e3 / 14.6959) # absolute upstream gas pressure (Pa)
P_psi = P_pa*(14.6959/101.325e3) #absolute upstream gas pressure (psi)
mu = visc_n2(T, P_pa) # viscosity of gas (Pa-s)
Gf = np.pi*D**2/(4*L) # geometric factor for axial flow in eqn 19 (m)

# check = P_psi / L_in; check
```

```python
n = len(time)-1 # array index size
zr = z_n2(Tstd,Pstd) # compressibility at reference temperature and pressure
rhor =  rho_n2(Tstd,Pstd) # gas density of reference temperature and pressure

rho = rho_n2(T,P_pa)# upstream gas density (kg/m3)
mdot = np.zeros(n) # mass flow rate (kg/sec)
yn = np.zeros(n) # flow rate function (m3/sec)
pg_pa = np.zeros(n) # geometric mean pressure (Pa)
Ppore_pa = np.zeros(n) # mean pore pressure (Pa)
qs =  np.zeros(n) # initial estimate of SS volumetric flow rate at upstream side of sample (m3/sec)
qr =  np.zeros(n) # volumetric flow rate at reference temperature and pressure
zm = np.zeros(n) 
kg =  np.zeros(n) #apparent permeability (m2)

for i in xrange(n):
    mdot[i] = Vt*(rho[i+1]-rho[i])/(time[i+1]-time[i]) # mass flow rate into sample (kg/sec)
    qs[i] = mdot[i] / rho[i+1] # approx SS vol. flow rate (m3/sec)
    qr[i] = rho[i+1]*qs[i] / rhor
    
#     yn[i] = Vt*math.log(p_pa[i]/p_pa[i+1])/(time[i+1]-time[i]) # [B-12], approximation of dp/dt
    pg_pa[i] = np.sqrt(p_pa[i]*p_pa[i+1]) #geometric mean of pressure [B-13], mean pressure at time[i]
    Ppore_pa[i] = 0.5*pg_pa[i]+Patm_pa
    zm[i] = z_n2(T[i+1],Ppore_pa[i])
    NUM = 2*c2*mu[i+1]*P*qr[i]*zm[i]
    DEN = c1*zr*Gf*(P_pa[i+1]-Patm_pa)*(P_pa[i+1]+Patm_pa)
    kg[i] = NUM/DEN #m2

print NUM
```

## Gas Slippage

\begin{equation}
    k_g = k_{\infty}\left(1+\frac{b}{P}\right)
\end{equation}
where:
- $k_g$ = apparent permeability of the medium to a particular gas (non-slip-corrected).
- $k_{\infty}$ = slip-corrected (Klinkenberg) permeability
- $b$ = Klinkenberg slip factor.
- $P$ = Absolute pressure.

\begin{equation}
    \bar{\lambda}=\frac{br}{4cP}
\end{equation}   

- $\bar{\lambda}$ = mean free path of gas molecules



- s = distance along the direction of flow.
- $q_s$ = volumetric rate of gas flow through area (A) per unit time. 
- A = area perpendicular to the flow of gas.
- $\frac{dP}{ds}$ = pressure gradient along s to which qs / A refers.
- μ = dynamic gas viscosity.

- β = coefficient of inertial resistivity.
- ρ = gas density.
