```python
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
def ss_pres(p_us, p_ds, length):
    """
    x = array of x position
    """
    x = np.linspace(0, length)  # inches
    term1 = (p_ds**2 - p_us**2)/length
    p = np.sqrt(term1*x + p_us**2)
    return p
```

```python
P_US = 20.0  # psi
P_DS = 5.0  # psi
LENGTH = 4.0  # inches
```

```python
DOMAIN = np.linspace(0, LENGTH)
PRES = ss_pres(P_US, P_DS, LENGTH)
```

```python
TERMA = (P_DS**2 - P_US**2)/LENGTH
TERMB = P_US**2
TERMC = TERMA*LENGTH + TERMB
P_AVG = (2/(3*TERMA)*(TERMC**1.5 - (TERMB**1.5)))/LENGTH
```

```python
P_AVG
```

```python
FIG_I, AX1 = plt.subplots(figsize=(12, 8))

LBL = ['Analytical Solution']
AX1.plot(DOMAIN, PRES, 'o-', lw=1)
AX1.legend(LBL, frameon=1, framealpha=1, loc=0)

AX1.set_xlabel('Length [in]', fontsize=12)
AX1.set_ylabel('Pressure [psi]', fontsize=12)
AX1.set_title(''r'One-Dimensional Steady-State Analysis', fontsize=16)
AX1.grid(b=True, which='major')
AX1.grid(b=True, which='major')
# plt.show()
# plt.close('all')
```
